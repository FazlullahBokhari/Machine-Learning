```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
#df = df[df['species'] != 'setosa']
```

```python
df['species'].unique()
```

```python
df['species'] = df['species'].map({'setosa':0,'versicolor':1,'virginica':2})
```

```python
## features into independent and dependent feature
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion='entropy')
classifier
```

```python
classifier.fit(X,y)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
classifier = DecisionTreeClassifier(criterion='gini')
classifier
```

```python
classifier.fit(X,y)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
classifier = DecisionTreeClassifier(criterion='gini',max_depth=2) # post pournning 
classifier
```

```python
classifier.fit(X,y)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
## pre orunning with gridsearchCV and cross valscore 
# Decision tree regressor
from sklearn.model_selection import GridSearchCV,train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)

param = {'max_depth':[2,4,6,8,10,12],
        'min_samples_split':[2,3,4],
        'min_samples_leaf':[1,2]}
clf = tree.DecisionTreeClassifier()
gcv = GridSearchCV(estimator=clf,param_grid=param)
gcv.fit(X_train,y_train)
```

```python
gcv.best_params_
```

```python
classifier = DecisionTreeClassifier(criterion='gini',max_depth=4,min_samples_split=2,min_samples_leaf=1) # post pournning 
classifier
```

```python
classifier.fit(X_test,y_test)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```
